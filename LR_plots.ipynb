```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./logs/log_CLR18.csv')
data_adv = pd.read_csv('./logs/log_AdversarialAE@03.06.2020-17:09:56.csv')
data_beta = pd.read_csv('./logs/log_BetaVAE@03.06.2020-17:14:45.csv')
data_van = pd.read_csv('./logs/log_VanillaVAE@03.06.2020-17:11:32.csv')
data_wass = pd.read_csv('./logs/log_WassersteinAE@03.06.2020-17:12:47.csv')
data_jig = pd.read_csv('./logs/log_JigsawVAE@06.06.2020-16:38:07.csv')
data_var = pd.read_csv('./logs/log_VarianceConstrainedAE@07.06.2020-09:31:07.csv')
```

```python
fig, axes = plt.subplots(figsize = (20, 7), nrows = 1, ncols = 2)
axes[0].tick_params(labelsize = 15, length = 5)
axes[0].plot(data['train_loss'], linewidth = 2)
axes[0].set_xlabel('# iterations', size = 16)
axes[0].set_ylabel('Train Loss', size = 16)
axes[1].tick_params(labelsize = 15, length = 5)
axes[1].plot(data['val_loss'], linewidth = 2)
axes[1].set_xlabel('# epochs', size = 16)
axes[1].set_ylabel('Val Loss', size = 16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.plot(data_adv['train_loss_mean'], label = 'AAE')
ax.plot(data_beta['train_loss_mean'], label = 'beta-AE')
ax.plot(data_van['train_loss_mean'], label = 'Vanilla VAE')
ax.plot(data_wass['train_loss_mean'], label = 'WAE')
ax.plot(data_jig['train_loss_mean'], label = 'Jigsaw AE')
ax.plot(data_var['train_loss_mean'], label = 'VCAE')
ax.set_ylim(30, 300)
ax.set_xlim(0, 3000)
ax.legend(loc = 'upper right', fontsize = 14)
ax.set_ylabel('Model Loss', size = 16)
ax.set_xlabel('# epochs', size = 16)
ax.tick_params(labelsize = 15, length = 5)
```

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.plot(data_adv['test_loss_mean'], label = 'AAE')
ax.plot(data_beta['test_loss_mean'], label = 'beta-AE')
ax.plot(data_van['test_loss_mean'], label = 'Vanilla VAE')
ax.plot(data_wass['test_loss_mean'], label = 'WAE')
ax.plot(data_jig['test_loss_mean'], label = 'Jigsaw AE')
ax.plot(data_var['test_loss_mean'], label = 'VCAE')
ax.set_ylim(30, 300)
ax.set_xlim(0, 3000)
ax.legend(loc = 'upper right', fontsize = 14)
ax.set_ylabel('Model Loss', size = 16)
ax.set_xlabel('# epochs', size = 16)
ax.tick_params(labelsize = 15, length = 5)
```
